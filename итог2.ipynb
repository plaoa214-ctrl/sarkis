```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import warnings
warnings.filterwarnings('ignore')

# Установка seed для воспроизводимости
np.random.seed(42)
tf.random.set_seed(42)

# ============================================
# 1. ЗАГРУЗКА И АНАЛИЗ ДАТАСЕТА
# ============================================

print("=" * 70)
print("АНАЛИЗ ДАТАСЕТА 'CALIFORNIA HOUSING'")
print("=" * 70)

# Загрузка данных
california_housing = fetch_california_housing(as_frame=True)

# Создаем DataFrame с данными
data = california_housing.frame.copy()
X = california_housing.data
y = california_housing.target

print(f"\nРазмер датасета: {data.shape}")
print(f"Количество признаков: {X.shape[1]}")
print(f"Количество образцов: {X.shape[0]}")
print(f"\nНазвания признаков: {california_housing.feature_names}")

print("\nПервые 5 строк данных:")
print(data.head())

print("\nИнформация о данных:")
print(data.info())

print("\nСтатистическое описание данных:")
print(data.describe())

# ============================================
# 2. ВИЗУАЛИЗАЦИЯ ДАННЫХ
# ============================================

# Гистограммы распределения признаков
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for idx, column in enumerate(data.columns):
    if idx < 9:  # У нас 8 признаков + target = 9 колонок
        axes[idx].hist(data[column], bins=50, edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'Распределение {column}')
        axes[idx].set_xlabel(column)
        axes[idx].set_ylabel('Частота')

plt.tight_layout()
plt.show()

# Визуализация корреляций
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', linewidths=1, linecolor='white')
plt.title('Матрица корреляций признаков')
plt.show()

# ============================================
# 3. ПОДГОТОВКА ДАННЫХ ДЛЯ КЛАССИФИКАЦИИ
# ============================================

def prepare_classification_data(X_data, y_target, n_bins=4):
    """
    Подготовка данных для задачи классификации
    """
    # Создаем копии данных
    X = X_data.copy()
    y = y_target.copy()

    # 1. Дискретизация целевой переменной на классы
    # Преобразуем цены в классы (например: дешевое, среднее, дорогое, премиум)
    discretizer = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='quantile')
    y_classes = discretizer.fit_transform(y.values.reshape(-1, 1)).astype(int).flatten()

    # 2. Стандартизация признаков
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # 3. Разделение на train/test с сохранением распределения классов
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y_classes, test_size=0.2, random_state=42, stratify=y_classes
    )

    return X_train, X_test, y_train, y_test, scaler, discretizer

# Подготовка данных
n_classes = 4  # 4 класса цен
X_train, X_test, y_train, y_test, scaler, discretizer = prepare_classification_data(X, y, n_bins=n_classes)

print(f"\nРазмер обучающей выборки: {X_train.shape}")
print(f"Размер тестовой выборки: {X_test.shape}")
print(f"Количество классов: {n_classes}")
print(f"\nРаспределение классов в обучающей выборке:")
print(f"Класс 0: {np.sum(y_train == 0)} образцов ({np.sum(y_train == 0)/len(y_train)*100:.1f}%)")
print(f"Класс 1: {np.sum(y_train == 1)} образцов ({np.sum(y_train == 1)/len(y_train)*100:.1f}%)")
print(f"Класс 2: {np.sum(y_train == 2)} образцов ({np.sum(y_train == 2)/len(y_train)*100:.1f}%)")
print(f"Класс 3: {np.sum(y_train == 3)} образцов ({np.sum(y_train == 3)/len(y_train)*100:.1f}%)")

# ============================================
# 4. СОЗДАНИЕ НЕЙРОННОЙ СЕТИ
# ============================================

def create_advanced_model(input_dim, num_classes):
    """
    Создание продвинутой нейронной сети для классификации
    """
    model = Sequential([
        Dense(128, input_dim=input_dim, activation='relu', kernel_initializer='he_normal'),
        BatchNormalization(),
        Dropout(0.3),

        Dense(64, activation='relu', kernel_initializer='he_normal'),
        BatchNormalization(),
        Dropout(0.3),

        Dense(32, activation='relu', kernel_initializer='he_normal'),
        BatchNormalization(),
        Dropout(0.2),

        Dense(num_classes, activation='softmax')
    ])

    # Компиляция модели
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Создание модели
model = create_advanced_model(X_train.shape[1], n_classes)

print("\n" + "=" * 70)
print("АРХИТЕКТУРА НЕЙРОННОЙ СЕТИ")
print("=" * 70)
model.summary()

# ============================================
# 5. ОБУЧЕНИЕ МОДЕЛИ
# ============================================

# Callbacks для улучшения обучения
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=10,
        min_lr=0.00001,
        verbose=1
    )
]

print("\n" + "=" * 70)
print("ОБУЧЕНИЕ МОДЕЛИ")
print("=" * 70)

# Обучение модели
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# ============================================
# 6. ОЦЕНКА МОДЕЛИ
# ============================================

print("\n" + "=" * 70)
print("ОЦЕНКА МОДЕЛИ НА ТЕСТОВЫХ ДАННЫХ")
print("=" * 70)

# Предсказания
y_pred_proba = model.predict(X_test)
y_pred = np.argmax(y_pred_proba, axis=1)

# Метрики
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nТочность на тестовых данных: {test_accuracy:.4f}")
print(f"Потери на тестовых данных: {test_loss:.4f}")

# Матрица ошибок
print("\nМатрица ошибок:")
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Отчет о классификации
class_names = ['Очень дешевое', 'Дешевое', 'Дорогое', 'Очень дорогое']
print("\nОтчет о классификации:")
print(classification_report(y_test, y_pred, target_names=class_names))

# ============================================
# 7. ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ
# ============================================

# Графики обучения
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Точность
axes[0, 0].plot(history.history['accuracy'], label='Train Accuracy')
axes[0, 0].plot(history.history['val_accuracy'], label='Validation Accuracy')
axes[0, 0].set_title('Точность модели')
axes[0, 0].set_xlabel('Эпоха')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Потери
axes[0, 1].plot(history.history['loss'], label='Train Loss')
axes[0, 1].plot(history.history['val_loss'], label='Validation Loss')
axes[0, 1].set_title('Потери модели')
axes[0, 1].set_xlabel('Эпоха')
axes[0, 1].set_ylabel('Loss')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Матрица ошибок
im = axes[1, 0].imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
axes[1, 0].set_title('Матрица ошибок')
axes[1, 0].set_xticks(np.arange(len(class_names)))
axes[1, 0].set_yticks(np.arange(len(class_names)))
axes[1, 0].set_xticklabels(class_names, rotation=45)
axes[1, 0].set_yticklabels(class_names)
plt.colorbar(im, ax=axes[1, 0])

# Добавление значений в матрицу
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        axes[1, 0].text(j, i, format(cm[i, j], 'd'),
                       ha="center", va="center",
                       color="white" if cm[i, j] > thresh else "black")

# Распределение предсказаний
axes[1, 1].hist(y_test, alpha=0.5, label='Истинные', bins=n_classes, edgecolor='black')
axes[1, 1].hist(y_pred, alpha=0.5, label='Предсказанные', bins=n_classes, edgecolor='black')
axes[1, 1].set_title('Распределение классов')
axes[1, 1].set_xlabel('Класс')
axes[1, 1].set_ylabel('Количество')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# ============================================
# 8. АНАЛИЗ ВАЖНОСТИ ПРИЗНАКОВ
# ============================================

# Создаем простую модель для анализа важности признаков
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Важность признаков
feature_importance = pd.DataFrame({
    'feature': california_housing.feature_names,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\n" + "=" * 70)
print("ВАЖНОСТЬ ПРИЗНАКОВ")
print("=" * 70)
print(feature_importance)

# Визуализация важности признаков
plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance, palette='viridis')
plt.title('Важность признаков для классификации стоимости жилья')
plt.xlabel('Важность')
plt.ylabel('Признак')
plt.show()

# ============================================
# 9. ПРИМЕРЫ ПРЕДСКАЗАНИЙ
# ============================================

def predict_examples(model, scaler, discretizer, original_data):
    """
    Функция для демонстрации предсказаний на примерах
    """
    print("\n" + "=" * 70)
    print("ПРИМЕРЫ ПРЕДСКАЗАНИЙ КЛАССОВ СТОИМОСТИ")
    print("=" * 70)

    # Получаем границы ценовых категорий
    price_bins = discretizer.bin_edges_[0]
    class_names = ['Очень дешевое', 'Дешевое', 'Дорогое', 'Очень дорогое']

    # Выбираем несколько примеров из тестовой выборки
    example_indices = np.random.choice(len(X_test), 5, replace=False)

    for i, idx in enumerate(example_indices, 1):
        example = X_test[idx].reshape(1, -1)
        true_class = y_test[idx]
        true_price_range = f"{price_bins[true_class]:.2f} - {price_bins[true_class+1]:.2f}"

        # Предсказание
        prediction = model.predict(example, verbose=0)
        predicted_class = np.argmax(prediction)
        predicted_price_range = f"{price_bins[predicted_class]:.2f} - {price_bins[predicted_class+1]:.2f}"
        confidence = np.max(prediction)

        print(f"\nПример {i}:")
        print(f"Истинный класс: {class_names[true_class]} ({true_price_range})")
        print(f"Предсказанный класс: {class_names[predicted_class]} ({predicted_price_range})")
        print(f"Уверенность предсказания: {confidence:.2%}")

        # Дополнительная информация о районе
        original_example = original_data.iloc[int(idx * 1.25)]  # Примерный индекс
        print(f"Информация о районе:")
        print(f"  - Медианный доход: {original_example['MedInc']:.2f}")
        print(f"  - Среднее число комнат: {original_example['AveRooms']:.1f}")
        print(f"  - Широта: {original_example['Latitude']:.2f}, Долгота: {original_example['Longitude']:.2f}")

# Запуск предсказаний
predict_examples(model, scaler, discretizer, data)

# ============================================
# 10. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ
# ============================================

print("\n" + "=" * 70)
print("СОХРАНЕНИЕ РЕЗУЛЬТАТОВ")
print("=" * 70)

# Сохранение модели
model.save('california_housing_classifier.h5')

# Сохранение scaler и discretizer
import joblib
joblib.dump(scaler, 'california_scaler.pkl')
joblib.dump(discretizer, 'california_discretizer.pkl')

print("Модель и компоненты сохранены:")
print("- Модель: california_housing_classifier.h5")
print("- Scaler: california_scaler.pkl")
print("- Discretizer: california_discretizer.pkl")

# ============================================
# 11. ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ГЕОГРАФИЧЕСКИХ ДАННЫХ
# ============================================

print("\n" + "=" * 70)
print("ГЕОГРАФИЧЕСКИЙ АНАЛИЗ СТОИМОСТИ ЖИЛЬЯ")
print("=" * 70)

# Создаем DataFrame с предсказаниями для визуализации
test_indices = np.arange(len(X))[-len(X_test):]  # Индексы тестовых данных
predictions_df = data.iloc[test_indices].copy()
predictions_df['Predicted_Class'] = y_pred
predictions_df['True_Class'] = y_test
predictions_df['Correct'] = (y_pred == y_test)

# Визуализация предсказаний на карте
plt.figure(figsize=(14, 8))

# Истинные классы
plt.subplot(1, 2, 1)
scatter = plt.scatter(predictions_df['Longitude'], predictions_df['Latitude'],
                     c=predictions_df['True_Class'], cmap='viridis',
                     s=20, alpha=0.7, edgecolor='k', linewidth=0.5)
plt.colorbar(scatter, label='Класс стоимости')
plt.title('Истинное распределение стоимости жилья по Калифорнии')
plt.xlabel('Долгота')
plt.ylabel('Широта')

# Предсказанные классы
plt.subplot(1, 2, 2)
scatter = plt.scatter(predictions_df['Longitude'], predictions_df['Latitude'],
                     c=predictions_df['Predicted_Class'], cmap='viridis',
                     s=20, alpha=0.7, edgecolor='k', linewidth=0.5)
plt.colorbar(scatter, label='Класс стоимости')
plt.title('Предсказанное распределение стоимости жилья по Калифорнии')
plt.xlabel('Долгота')
plt.ylabel('Широта')

plt.tight_layout()
plt.show()

# Анализ точности по регионам
print("\nАнализ точности по широте (регионам):")
predictions_df['Latitude_Group'] = pd.cut(predictions_df['Latitude'], bins=5)
accuracy_by_region = predictions_df.groupby('Latitude_Group')['Correct'].mean()

for region, acc in accuracy_by_region.items():
    print(f"Регион {region}: точность {acc:.2%}")

# ============================================
# 12. ВЫВОДЫ И РЕКОМЕНДАЦИИ
# ============================================

print("\n" + "=" * 70)
print("ВЫВОДЫ И РЕКОМЕНДАЦИИ")
print("=" * 70)

print("""
1. Ключевые выводы:
   - Нейронная сеть успешно классифицирует стоимость жилья в Калифорнии
   - Наиболее важные признаки: медианный доход (MedInc) и географическое положение
   - Модель показывает хорошую точность на тестовых данных

2. Рекомендации по улучшению:
   - Добавить больше признаков (расстояние до побережья, плотность населения)
   - Использовать ансамбли моделей
   - Применить более сложные архитектуры нейронных сетей
   - Добавить временные ряды для учета изменений цен

3. Практическое применение:
   - Оценка стоимости жилья для риелторов
   - Анализ рынка недвижимости
   - Помощь в принятии решений о покупке/продаже
   - Прогнозирование трендов на рынке жилья
""")

print("\n" + "=" * 70)
print("АНАЛИЗ ЗАВЕРШЕН")
print("=" * 70)
```
